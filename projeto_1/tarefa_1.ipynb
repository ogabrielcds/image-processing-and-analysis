```python
import matplotlib.pyplot as plt
from PIL import Image
import requests
import numpy as np
```

```python
# Abrir e visualizar a imagem escolhida:
url = "https://www.daninoce.com.br/wp-content/uploads/2020/05/tipos-de-maa-destaque-960x625.jpg"

image = Image.open(requests.get(url, stream=True).raw)
print(f"A imagem tem o modo: {image.mode}")

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.imshow(image)
plt.show()
```

```python
# Converter a imagem para monocromática
gray_image = image.convert("L")
print(f"A imagem tem o modo: {gray_image.mode}")

# Exibir a imagem:
fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.imshow(gray_image, cmap="gray")
plt.show()
```

```python
# Dimensão da imagem:
largura, altura = gray_image.size
print(f'Largura: {largura} pixels')
print(f'Altura: {altura} pixels')
```

```python
# Máximo e mínimo
img_array = np.array(gray_image)
min_val = np.min(img_array)
max_val = np.max(img_array)

print(f'O valor mínimo é {min_val}')
print(f'O valor máximo é {max_val}')
```

```python
# Exibir a linha:
linha_meio = img_array[altura//2]

plt.plot(linha_meio)
plt.xlabel("Posição")
plt.ylabel("Intensidade")
plt.show()
```

```python
# Exibir histograma
plt.hist(linha_meio, bins=256)
plt.xlabel("Intensidade")
plt.ylabel("Frequência")
plt.title("Histograma da linha do meio")
plt.show()
```
